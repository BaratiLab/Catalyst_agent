```python
import ase.io as io
from ase.spacegroup import get_spacegroup

# Replace 'path/to/your/material_file' with the actual path to your file
# This could be a POSCAR, CIF, XYZ, or other supported format.
file_path = '/home/ach/Documents/cathub/co2rr_cifs/Cu36.cif' 

# Read the material file into an Atoms object
atoms = io.read(file_path)

# Get the Spacegroup object
# symprec is a tolerance for symmetry detection, adjust if needed
spacegroup = get_spacegroup(atoms, symprec=1e-2)

# Print the space group number and symbol
print(f"Space group number: {spacegroup.no}")
print(f"Space group symbol: {spacegroup.symbol}")

# You can also access other properties of the Spacegroup object
# For example, the scaled primitive cell or symmetry operations
# print(spacegroup.scaled_primitive_cell)
# print(spacegroup.get_rotations())
```
